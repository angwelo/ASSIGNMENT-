```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Data centers (rectangles)
ax.add_patch(patches.Rectangle((0.5, 3.5), 2, 1, edgecolor='black', facecolor='#d9edf7'))
ax.text(1.5, 4, "Data Center A\n(Nairobi)", ha='center', va='center')

ax.add_patch(patches.Rectangle((4.5, 3.5), 2, 1, edgecolor='black', facecolor='#d9edf7'))
ax.text(5.5, 4, "Data Center B\n(Mombasa)", ha='center', va='center')

# Global DNS Load Balancer
ax.add_patch(patches.Rectangle((2.5, 5), 2, 1, edgecolor='black', facecolor='#fcf8e3'))
ax.text(3.5, 5.5, "DNS / Anycast\nGlobal Load Balancer", ha='center', va='center')

# Hardware Load Balancers
ax.add_patch(patches.Rectangle((0.7, 2.2), 1.6, 0.8, edgecolor='black', facecolor='#f2dede'))
ax.text(1.5, 2.6, "Hardware LB\n(F5/Huawei)", ha='center', va='center')

ax.add_patch(patches.Rectangle((4.7, 2.2), 1.6, 0.8, edgecolor='black', facecolor='#f2dede'))
ax.text(5.5, 2.6, "Hardware LB\n(F5/Huawei)", ha='center', va='center')

# Application/API Load Balancers
ax.add_patch(patches.Rectangle((0.9, 1), 1.2, 0.6, edgecolor='black', facecolor='#dff0d8'))
ax.text(1.5, 1.3, "API Gateway\n(Nginx/HAProxy)", ha='center', va='center')

ax.add_patch(patches.Rectangle((4.9, 1), 1.2, 0.6, edgecolor='black', facecolor='#dff0d8'))
ax.text(5.5, 1.3, "API Gateway\n(Nginx/HAProxy)", ha='center', va='center')

# Users at bottom
ax.add_patch(patches.Rectangle((2.5, 0), 2, 0.7, edgecolor='black', facecolor='#eeeeee'))
ax.text(3.5, 0.35, "Safaricom Users\n(M-Pesa, Calls, Data)", ha='center', va='center')

# Arrows from DNS to data centers
ax.annotate("", xy=(1.5, 3.5), xytext=(3.5, 5),
            arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(5.5, 3.5), xytext=(3.5, 5),
            arrowprops=dict(arrowstyle="->", lw=2))

# Arrows inside data centers
ax.annotate("", xy=(1.5, 2.2), xytext=(1.5, 3.5),
            arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(5.5, 2.2), xytext=(5.5, 3.5),
            arrowprops=dict(arrowstyle="->", lw=2))

# Arrows to API Gateways
ax.annotate("", xy=(1.5, 1), xytext=(1.5, 2.2),
            arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(5.5, 1), xytext=(5.5, 2.2),
            arrowprops=dict(arrowstyle="->", lw=2))

# Arrows from users to both API gateways
ax.annotate("", xy=(1.5, 1), xytext=(3.5, 0.7),
            arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(5.5, 1), xytext=(3.5, 0.7),
            arrowprops=dict(arrowstyle="->", lw=2))

# Hide axes
ax.axis('off')

plt.title("Safaricom Load Balancing Architecture (Simplified)", fontsize=14, weight='bold')
plt.show()
```
